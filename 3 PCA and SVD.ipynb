### LU Decomposition

```python
import numpy as np
from scipy.linalg import lu
```

```python
m = np.array([[4,5,6],[2,1,4],[9,1,3]])
m
```

```python
P, L, U = lu(m)
```

```python
print(L)
```

```python
print(U)
```

```python
print(P)
```

```python
P @ L @ U
```

### Principal Component Analysis

```python
data = np.array([[67,3,1089,56],[48,1,1234,78], [87,7,1256,59],
               [76,5,1052,58],[71,9,1114,62],[71,6,1068,50],
               [73,8,1139,49],[70,4,1139,60]])
```

```python
data
```

```python
data.shape
```

```python
#import the class
from sklearn.decomposition import PCA

#create the object
pca = PCA()

x_scaled = scaler.fit_transform(data)


#apply the algorithm
x_pca = pca.fit_transform(x_scaled)
```

```python
x_pca
```

```python
pca.explained_variance_ratio_
```

```python
plt.xlabel('Feature Number')
plt.ylabel('Variance')
plt.grid()
plt.bar(range(4), pca.explained_variance_ratio_, color = 'black')
plt.plot(range(4), pca.explained_variance_ratio_, marker ='.',color='red')
```

```python
x_pca.shape
```

```python
pca = PCA(n_components = 2)
```

```python
x_pca = pca.fit_transform(x_scaled)
x_pca
```
